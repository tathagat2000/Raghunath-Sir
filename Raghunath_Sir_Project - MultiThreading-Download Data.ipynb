```python
import requests
import pandas as pd
from io import BytesIO
import zipfile
import io
import os
import datetime
from time import time
from concurrent.futures import ThreadPoolExecutor, as_completed , ProcessPoolExecutor
```

```python
def MonthCode(m):
    if m == 1:
        return "JAN"
    elif m == 2:
        return "FEB"
    elif m == 3:
        return "MAR"
    elif m == 4:
        return "APR"
    elif m == 5:
        return "MAY"
    elif m == 6:
        return "JUN"
    elif m == 7:
        return "JUL"
    elif m == 8:
        return "AUG"
    elif m == 9:
        return "SEP"
    elif m == 10:
        return "OCT"
    elif m == 11:
        return "NOV"
    elif m == 12:
        return "DEC"
```

```python
def twoDigits(day) :
    if len(day) == 2 :
        return day
    else :
        day = '0' + day
        return day
```

```python
def makeListOfUrls(start_date , end_date) :
    allUrls = []
    delta = datetime.timedelta(days=1)
    constant_url = "https://archives.nseindia.com/content/historical/DERIVATIVES/"
    current_date = start_date
    while current_date <= end_date :
        variable_url = str(current_date.year) + '/' + MonthCode(current_date.month) + '/' + "fo" + twoDigits(str(current_date.day)) + MonthCode(current_date.month) + str(current_date.year) + "bhav.csv.zip"
        url = constant_url + variable_url
        allUrls.append(url)
        current_date = current_date + delta
        while ((current_date.weekday() == 5) or (current_date.weekday()==6)) :
            current_date = current_date + delta
    return allUrls
    
```

```python
def downloadData(url) :
    temp = url.find('fo')
    temp1 = url[temp:]
    zipped_filename = r"C:\Users\Tathagat\Desktop\BhavCopy\Zippped_Bhavcopy" + "\\" + temp1
    unzipped_filename = r"C:\Users\Tathagat\Desktop\BhavCopy\UnZipped_CSV_Bhavcopy" 
    flag=False
    try : 
        req = requests.get(url)
        print("Got Data " , temp1)
        flag=True

    except :
        print("No Data for selected Date " , temp1)

    if flag :
        file = open(zipped_filename , 'wb')
        for chunk in req.iter_content(100000):
            file.write(chunk)
        file.close()

        with zipfile.ZipFile(zipped_filename, 'r') as zip_ref:
            zip_ref.extractall(unzipped_filename)




      
```

```python

start_date = datetime.datetime(2020,1,1)
end_date = datetime.datetime.now()
start = time()
allUrls = makeListOfUrls(start_date , end_date)
processes = []
with ThreadPoolExecutor(max_workers=16) as executor:
    for url in allUrls:
        processes.append(executor.submit(downloadData, url))
        

    
print(f'Time taken: {time() - start}')

```
