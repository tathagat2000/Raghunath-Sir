```python
import requests
import pandas as pd
from io import BytesIO
import zipfile
import io
import os
import datetime
from pathlib import Path
import numpy as np
```

```python
def MonthCode(m):
    if m == 1:
        return "JAN"
    elif m == 2:
        return "FEB"
    elif m == 3:
        return "MAR"
    elif m == 4:
        return "APR"
    elif m == 5:
        return "MAY"
    elif m == 6:
        return "JUN"
    elif m == 7:
        return "JUL"
    elif m == 8:
        return "AUG"
    elif m == 9:
        return "SEP"
    elif m == 10:
        return "OCT"
    elif m == 11:
        return "NOV"
    elif m == 12:
        return "DEC"
```

```python
def twoDigits(day) :
    if len(day) == 2 :
        return day
    else :
        day = '0' + day
        return day
```

```python
def variablePath(currentDate) :
    path = "fo" + twoDigits(str(currentDate.day)) + MonthCode(currentDate.month) + str(currentDate.year) + "bhav.csv"
    return path
```

```python
def operate(path) :
    stockPrice = pd.read_csv(path)
    stockPrice  = stockPrice[((stockPrice['INSTRUMENT']=="OPTIDX") | (stockPrice['INSTRUMENT']=="OPTSTK"))]
    stockVolume = {}
    for idx in stockPrice.index :
        symbol = stockPrice.loc[idx , 'SYMBOL']
        volume = stockPrice.loc[idx , 'CONTRACTS']
        if symbol in stockVolume.keys() :
            stockVolume[symbol]+=volume
        else :
            stockVolume[symbol] = volume
            
    return stockVolume
```

```python
startDate = datetime.datetime(2020,9,24)
endDate = datetime.datetime.now()
currentDate = startDate
delta = datetime.timedelta(days=1)
dateVsStock={}
while currentDate<=endDate :
    path = pathToFiles + "\\" + variablePath(currentDate)
    currentFile = Path(path)
    if currentFile.is_file() :
        print(currentDate.date())
        stockVolume = operate(path)
        dateVsStock[currentDate.date()] = stockVolume 
    
    currentDate = currentDate + delta
    while ((currentDate.weekday() == 5) or (currentDate.weekday()==6)) :
            currentDate = currentDate + delta


dateVsStockDf = pd.DataFrame(dateVsStock)
dateVsStockDf = dateVsStockDf.T
dateVsStockDf = dateVsStockDf.reindex(sorted(dateVsStockDf.columns), axis=1)
dateVsStockDf.to_csv(csvPath) 
    
```
