```python
import requests
import pandas as pd
from io import BytesIO
import zipfile
import io
import os
import datetime
from pathlib import Path
import numpy as np
```

```python
def MonthCode(m):
    if m == 1:
        return "JAN"
    elif m == 2:
        return "FEB"
    elif m == 3:
        return "MAR"
    elif m == 4:
        return "APR"
    elif m == 5:
        return "MAY"
    elif m == 6:
        return "JUN"
    elif m == 7:
        return "JUL"
    elif m == 8:
        return "AUG"
    elif m == 9:
        return "SEP"
    elif m == 10:
        return "OCT"
    elif m == 11:
        return "NOV"
    elif m == 12:
        return "DEC"
```

```python
def twoDigits(day) :
    if len(day) == 2 :
        return day
    else :
        day = '0' + day
        return day
```

```python
def variablePath(currentDate) :
    path = "fo" + twoDigits(str(currentDate.day)) + MonthCode(currentDate.month) + str(currentDate.year) + "bhav.csv"
    return path
```

```python
pathToFiles = r"C:\Users\Tathagat\Desktop\BhavCopy\UnZipped_CSV_Bhavcopy"
```

```python
def operate(path,multiplier) :
    stockPrice = pd.read_csv(path)
    stockPrice  = stockPrice[((stockPrice['INSTRUMENT']=="OPTIDX") | (stockPrice['INSTRUMENT']=="OPTSTK"))]
    indicator = stockPrice[stockPrice['CONTRACTS'] > stockPrice['OPEN_INT']*multiplier]
    return indicator
    
```

```python
startDate = datetime.datetime(2020,1,1)
endDate = datetime.datetime.now()
currentDate = startDate
delta = datetime.timedelta(days=1)
dateVsDataFrame={}
multiplier=10
while currentDate<=endDate :
    path = pathToFiles + "\\" + variablePath(currentDate)
    currentFile = Path(path)
    if currentFile.is_file() :
        print(currentDate.date())
        indicator = operate(path,multiplier)
        dateVsDataFrame[currentDate.date()] = indicator
    
    currentDate = currentDate + delta
    while ((currentDate.weekday() == 5) or (currentDate.weekday()==6)) :
            currentDate = currentDate + delta

allIndicators = pd.concat(dateVsDataFrame.values() , ignore_index=True)
```
